#Shift operators

```
1. Left shift operator(<<) 
  - right hand side vaccant cells fill with 0
  - n cells from left side removed & n cells from right side added 
---------------------------------
print(10<<2) 
   0000....01010--10(binary 32bit)
   --00....0101000 (2cells from left side removed & 
                    2 cells from right side added here)

  00....0101000
  ...(2^5)+(2^4)+(2^3)+(2^2)+(2^1)+(2^0)
  =32+0+8+0+0+0
  =40
```

image.png




````
print(10<<2)
````



```
2. Right shift operator(>>)

-> left hand side vacant cells will be filled with sign bit.
-> sign bit for +ive number--0
-> sign bit for -ive number--1

------------------------
print(10>>2)

   0000....01010--10(binary 32bit)
 000000....010-- 2cells from right side removed & 
                    2 cells from added side added which filled with 
                    sign bit according to nunmber(10)
 
   0000....00010
  ...(2^5)+(2^4)+(2^3)+(2^2)+(2^1)+(2^0)
  =0+0+0+0+2+0
  =2




```
